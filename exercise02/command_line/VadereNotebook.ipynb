# Setting Pedestrians

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# function for reading postvis file and grouping results by id of pedestrians
def read_file(filename):
    data = pd.read_csv(filename, delimiter=" ")
    return data.groupby("pedestrianId")
```

```python
# the exact path to postvis file(SHOULD BE CHANGED WHEN TESTING)
postvis_path = "C:\\Users\\eminm\\OneDrive\\Desktop\\vadere.master.windows\\Scenarios\\TestScenarios\\CornerScenario\\output\\rimea_06_corner_self_osm_with_pedestrian_2021-05-13_01-39-03.800\\postvis.traj"
# read simulation data
ped = read_file(postvis_path)
ped_array = [pedestrian for pedestrian in ped]
# get information of simulation of new pedestrian with ID 2
new_pedestrian_info = ped_array[0][1]
# get the coordinates through which new pedestrian simulation iterates
starting_positions_pedestrianX = new_pedestrian_info["startX-PID1"]

starting_positions_pedestrianY = new_pedestrian_info["startY-PID1"]

# get time when new pedestrian reaches target
new_pedestrian_endtime = new_pedestrian_info["endTime-PID1"].array[max(len(starting_positions_pedestrianX), len(starting_positions_pedestrianY))-1]

#plot the simulation of the new pedestrian
plt.plot(new_pedestrian_info["endTime-PID1"], new_pedestrian_info["startX-PID1"])
plt.plot(new_pedestrian_info["endTime-PID1"], new_pedestrian_info["startY-PID1"])
plt.xlabel('Simulation Time')
plt.ylabel('X(blue) and Y(orange) coordinates')
plt.title('New Pedestrian simulation')
plt.savefig("new_pedestrian_coordinates_simulation.pdf")
plt.show()

###############################################
# get information of simulation of pedestrian from source
pedestrian_from_source_info = ped_array[1][1]
# get the coordinates through which source pedestrian simulation iterates
starting_positions_pedestrianX = pedestrian_from_source_info["startX-PID1"]

starting_positions_pedestrianY = pedestrian_from_source_info["startY-PID1"]
# get time when pedestrian from the source reaches target
source_pedestrian_endtime = pedestrian_from_source_info["endTime-PID1"].array[max(len(starting_positions_pedestrianX), len(starting_positions_pedestrianY))-1]

#plot the simulation of the source pedestrian
plt.plot(pedestrian_from_source_info["endTime-PID1"], pedestrian_from_source_info["startX-PID1"])
plt.plot(pedestrian_from_source_info["endTime-PID1"], pedestrian_from_source_info["startY-PID1"])
plt.xlabel('Simulation Time')
plt.ylabel('X(blue) and Y(orange) coordinates')
plt.title('Source Pedestrian simulation')
plt.savefig("source_pedestrian_coordinates_simulation.pdf")
plt.show()
```

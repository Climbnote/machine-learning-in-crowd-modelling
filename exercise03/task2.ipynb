```python
import numpy as np
import matplotlib.pyplot as plt
import visualization
```

## Task 2

```python
# define function for the first dynamical system: dot(x) = alpha - x^2
def fun1(x, y, alpha):
    u = x
    v = alpha - y**2
    return u, v
# define function for the second dynamical system: dot(x) = aplha - 2*x^2 - 4
def fun2(x, y, alpha):
    u = x
    v = alpha - 2* y**2 - 3
    return u, v
```

```python
# Different alphas showing different behavior
alphas1 = [-0.5, 0.0, 0.5]
# Start points for the trajectories we plot
starts = [
    [[-0.25,0.0]],
    [[0.1, 0.25]],
    [[-0.1, -0.25]]]
for i in range(len(alphas1)):
    # visualize plot for the first equation diagram phase
    plot = visualization.phase_diagram_with_trajectory_2d(fun1,alphas1[i],starts[1],arrows=False)
    plot.gca().set_title("$$ = {:.1f}".format(alphas1[i]))
    # save plot as file
    plt.savefig("task2/system1_phase_diagram_{i}.pdf".format(i=i))
    # show plot
    plt.show()
```

```python
# Different alphas showing different behavior
alphas2 = [1.0, 3.0, 3.5]
# Start points for the trajectories we plot
starts = [
    [[-0.25,0.0]],
    [[0.1, 0.25]],
    [[-0.1, -0.25]]]
for i in range(len(alphas1)):
    # visualize plot for the second equation diagram phase
    plot = visualization.phase_diagram_with_trajectory_2d(fun2,alphas2[i],starts[i],arrows=False)
    plot.gca().set_title("$$ = {:.1f}".format(alphas2[i]))
    # save plot as file
    plt.savefig("task2/system2_phase_diagram_{i}.pdf".format(i=i))
    # show plot
    plt.show()
```

```python
# visualize bifurcation diagram for the first dynamical system: dot(x) = alpha - x^2
def f(alpha, x):
    return alpha -  x**2

def f_x(alpha, x):
    return - 2.*x

plot = visualization.bifurcation_diagram_2d(f,f_x)

plot.savefig("task2/bifurcation_system1.pdf")
```

```python
# visualize bifurcation diagram for the first dynamical system: dot(x) = alpha - 2*x^2 - 3
def f(alpha, x):
    return alpha -  2 * x**2 - 3

def f_x(alpha, x):
    return - 4 * x

plot = visualization.bifurcation_diagram_2d(f,f_x,alpha_lim=(2.0,4.0))

plot.savefig("task2/bifurcation_system2.pdf")
```

```python
# visualize phase diagram for both system for alpha = 1.0
alpha_plus1 = 1.0
# plot phase diagram for the first system
plot = visualization.phase_diagram_with_trajectory_2d(fun1,alpha_plus1,[[0.1, 0.0]],arrows=False)
plot.gca().set_title("$$ = {:.1f}".format(alpha_plus1))
# save plot as file
plt.savefig("task2/system1_phase_diagram_alpha_plus1.pdf")
# show plot
plt.show()
# plot phase diagram for the second system
plot = visualization.phase_diagram_with_trajectory_2d(fun2,alpha_plus1,[[0.1, 0.0]],arrows=False)
plot.gca().set_title("$$ = {:.1f}".format(alpha_plus1))
# save plot as file
plt.savefig("task2/system2_phase_diagram_alpha_plus1.pdf")
# show plot
plt.show()
```

```python
# visualize phase diagram for both system for alpha = -1.0
alpha_minus1 = -1.0
# plot phase diagram for the first system
plot = visualization.phase_diagram_with_trajectory_2d(fun1,alpha_minus1,[[0.1, 0.0]],arrows=False)
plot.gca().set_title("$$ = {:.1f}".format(alpha_minus1))
# save plot as file
plt.savefig("task2/system1_phase_diagram_alpha_minus1.pdf")
# show plot
plt.show()
# plot phase diagram for the second system
plot = visualization.phase_diagram_with_trajectory_2d(fun2,alpha_minus1,[[0.1, 0.0]],arrows=False)
plot.gca().set_title("$$ = {:.1f}".format(alpha_minus1))
# save plot as file
plt.savefig("task2/system2_phase_diagram_alpha_minus1.pdf")
# show plot
plt.show()
```

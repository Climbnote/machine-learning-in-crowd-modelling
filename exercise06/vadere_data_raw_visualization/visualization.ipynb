```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
speed_bott_070 = np.loadtxt('../vadere_data_raw/bottleneck/bottleneck_070/areaSpeed.txt',skiprows=1)
density_bott_070 = np.loadtxt('../vadere_data_raw/bottleneck/bottleneck_070/areaDensity.txt',skiprows=1)
speed_bott_095 = np.loadtxt('../vadere_data_raw/bottleneck/bottleneck_095/areaSpeed.txt',skiprows=1)
density_bott_095 = np.loadtxt('../vadere_data_raw/bottleneck/bottleneck_095/areaDensity.txt',skiprows=1)
speed_bott_120 = np.loadtxt('../vadere_data_raw/bottleneck/bottleneck_120/areaSpeed.txt',skiprows=1)
density_bott_120 = np.loadtxt('../vadere_data_raw/bottleneck/bottleneck_120/areaDensity.txt',skiprows=1)
speed_bott_180 = np.loadtxt('../vadere_data_raw/bottleneck/bottleneck_180/areaSpeed.txt',skiprows=1)
density_bott_180 = np.loadtxt('../vadere_data_raw/bottleneck/bottleneck_180/areaDensity.txt',skiprows=1)
```

```python
#print the bottleneck scenario output files areaSpeed and areaDensity
time = np.arange(1,len(density_bott_070) + 1,1)

fig, ax = plt.subplots(2,4,figsize=(16,4), sharex=True)
ax[0][0].plot(time, speed_bott_070[:,1])
ax[0][0].set_ylim(0,1.5)
ax[0][0].set_title("w=0.70")
ax[0][0].set_ylabel("$v_{mean}$ in $m/s$") 
ax[1][0].plot(time, density_bott_070[:,1] / 10.8, c='red')
ax[1][0].set_ylim(0,5)
ax[1][0].set_ylabel("$\delta_{mean}$ in $ped/m^2$") 
ax[0][1].plot(time, speed_bott_095[:,1])
ax[0][1].set_ylim(0,1.5)
ax[0][1].set_title("$w=0.95$")
ax[1][1].plot(time, density_bott_095[:,1] / 10.8, c='red')
ax[1][1].set_ylim(0,5)
ax[0][2].plot(time, speed_bott_120[:,1])
ax[0][2].set_ylim(0,1.5)
ax[0][2].set_title("$w=1.20$")
ax[1][2].plot(time, density_bott_120[:,1] / 10.8, c='red')
ax[1][2].set_ylim(0,5)
ax[0][3].plot(time, speed_bott_180[:,1])
ax[0][3].set_ylim(0,1.5)
ax[0][3].set_title("$w=1.80$")
ax[1][3].plot(time, density_bott_180[:,1] / 10.8, c='red')
ax[1][3].set_ylim(0,5)
plt.show()
```

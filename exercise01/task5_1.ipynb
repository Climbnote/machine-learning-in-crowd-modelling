## Task 5/5: Tests

### TEST1:  RiMEA scenario 1 (straight line, ignore premovement time)

```python
#create scenario
corridor_scenario = scenario_reader.make_scenario('scenarios/corridor_scenario.json')

#visualize initial scenario
corridor_plot = visualization.plot_automaton_states(corridor_scenario.states)

#save plot
corridor_plot.savefig("results/task5_corridor.pdf")

#simulate the corridor scenario using the Automaton class
corridor_automaton = Automaton(corridor_scenario)
corridor_data = corridor_automaton.simulate_pedestrians()

#print travel time
corridor_cell_width = corridor_scenario.parameters.cell_width
corridor_speed_pedestrian = corridor_scenario.parameters.speed_pedestrian
print('The travel time of the pedestrian is', round(corridor_cell_width / corridor_speed_pedestrian * len(corridor_data), 3), 'seconds.')

#animate the simulated data using the Visualization class
%matplotlib notebook
corridor_animation = visualization.plot_automaton_simulation(corridor_data, Parameters(speed_pedestrian=1.33))
HTML(corridor_animation.to_jshtml())
```

```python
#save the animation as a video file
corridor_animation.save('results/task5_corridor.mp4', codec='h264')
```

```python
#save end state for report
corridor_plot_end = visualization.plot_automaton_states(corridor_data[-1])
corridor_plot_end.savefig("results/task5_corridor_end.pdf")
```

```python
from automaton import Scenario
from automaton import Automaton
from automaton import Parameters
import visualization
import numpy as np
from enum import Enum
from IPython.display import HTML
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib import animation
import math
import random
import scenario_reader
```

## Task 1/5: Setting up the modeling environment

### 1: Basic Visualization

```python
#to showcase the setup, a simple scenario is created and visualized

#scenario creation using scenario reader
basic_scenario = scenario_reader.make_scenario('scenarios/basic_scenario.json')

#plot the scenario using the Visualization class
basic_plot = visualization.plot_automaton_states(basic_scenario.states)

#save plot
basic_plot.savefig("results/task1_basic_scenario.pdf")
```

```python
#simulate the single scenario using the Automaton class
basic_automaton = Automaton(basic_scenario)
basic_data = basic_automaton.simulate_pedestrians()
```

```python
#animate the simulated data using the Visualization class

#set matplotlib environment
%matplotlib notebook
#call animation function and convert animation to JavaScript to be able to display it inline
basic_animation = visualization.plot_automaton_simulation(basic_data)
HTML(basic_animation.to_jshtml())
```

```python
#save the animation as a video file
basic_animation.save('results/task1_basic_scenario.mp4', codec='h264')
```

### 2: Adding Rectangular Obstacles

```python
#showcase rectangular adding of obstacles
rectangular_obstacle_scenario = Scenario()
rectangular_obstacle_scenario.add_obstacle_rectangular(2,2,5,5)
rectangular_obstacle_scenario.set_target(7,2)

#plot the scenario using the Visualization class
rectangular_obstacle_plot = visualization.plot_automaton_states(rectangular_obstacle_scenario.states)
```

```python
#save plot
rectangular_obstacle_plot.savefig("results/task1_add_obstacles_rectangular.pdf")
```

### 3: Adding pedestrians for density

```python
#showcase adding pedestrians for density
density_scenario = Scenario()
density_scenario.set_target(7,2)
#add 1P/m^2 in an area of 9m^2:
density_scenario.add_pedestrians_for_density(1, 1, 1, 6, 6)

#plot the scenario using the Visualization class
density_plot = visualization.plot_automaton_states(density_scenario.states)
```

```python
#save plot
density_plot.savefig("results/task1_add_pedestrians_for_density.pdf")
```

### TEST3:  RiMEA scenario 6 (movement around a corner)

```python
#create scenario
corner_scenario = scenario_reader.make_scenario('scenarios/corner_scenario.json')

#visualize initial scenario
corner_plot = visualization.plot_automaton_states(corner_scenario.states)

#save plot
corner_plot.savefig("results/task5_corner.pdf")

#simulate the corridor scenario using the Automaton class
corner_automaton = Automaton(corner_scenario)
corner_data = corner_automaton.simulate_pedestrians()


#animate the simulated data using the Visualization class
%matplotlib notebook
corner_animation = visualization.plot_automaton_simulation(corner_data, Parameters(speed_pedestrian=1.0))
HTML(corner_animation.to_jshtml())
```

```python
#save end for report
corner_plot_end = visualization.plot_automaton_states(corner_data[-1])

#save plot
corner_plot_end.savefig("results/task5_corner_timestep_end.pdf")
```

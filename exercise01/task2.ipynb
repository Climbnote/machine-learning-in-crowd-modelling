```python
from automaton import Scenario
from automaton import Automaton
from automaton import Parameters
import visualization
import numpy as np
from enum import Enum
from IPython.display import HTML
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib import animation
import math
import random
import scenario_reader
```

## Task 2/5: First step of a single pedestrian

```python
#create scenario
single_scenario = scenario_reader.make_scenario('scenarios/single_scenario.json')

#plot the scenario
single_plot = visualization.plot_automaton_states(single_scenario.states)

#save plot
single_plot.savefig("results/task2_single_scenario.pdf")
```

```python
#simulate the single scenario using the Automaton class
single_automaton = Automaton(single_scenario)
single_data = single_automaton.simulate_pedestrians()
```

```python
#animate the simulated data using the Visualization class

#set matplotlib environment
%matplotlib notebook
#call animation function and convert animation to JavaScript to be able to display it inline
single_animation = visualization.plot_automaton_simulation(single_data, Parameters(speed_pedestrian=1.0))
HTML(single_animation.to_jshtml())
```

```python
#save the animation as a video file
single_animation.save('results/task2_single_scenario.mp4', codec='h264')
```

```python
#save also endstate as figure for report
#plot the scenario
single_plot_end = visualization.plot_automaton_states(single_data[-1])

#save plot
single_plot_end.savefig("results/task2_single_scenario_end.pdf")
```

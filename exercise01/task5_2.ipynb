```python
#measure speed for density 0.5
measurement_fundamental_automaton1 = Automaton(measurement_fundamental_scenario1, measure_speed=True)
measurement_fundamental_automaton1.set_measuring_point(60, 63, 0.5, 1.33)
measurement_fundamental_data1 = measurement_fundamental_automaton1.simulate_pedestrians()

#measure speed for density 1.0
measurement_fundamental_automaton2 = Automaton(measurement_fundamental_scenario2, measure_speed=True)
measurement_fundamental_automaton2.set_measuring_point(60, 63, 1.0, 1.33)
measurement_fundamental_data2 = measurement_fundamental_automaton2.simulate_pedestrians()

#measure speed for density 2.0
measurement_fundamental_automaton3 = Automaton(measurement_fundamental_scenario3, measure_speed=True)
measurement_fundamental_automaton3.set_measuring_point(60, 63, 2.0, 1.33)
measurement_fundamental_data3 = measurement_fundamental_automaton3.simulate_pedestrians()

#measure speed for density 3.0
measurement_fundamental_automaton4 = Automaton(measurement_fundamental_scenario4, measure_speed=True)
measurement_fundamental_automaton4.set_measuring_point(60, 63, 3.0, 1.33)
measurement_fundamental_data4 = measurement_fundamental_automaton4.simulate_pedestrians()

#measure speed for density 4.0
measurement_fundamental_automaton5 = Automaton(measurement_fundamental_scenario5, measure_speed=True)
measurement_fundamental_automaton5.set_measuring_point(60, 63, 4.0, 1.33)
measurement_fundamental_data5 = measurement_fundamental_automaton5.simulate_pedestrians()

flows_array = np.array([measurement_fundamental_automaton1.average_speed * 0.5, 
                        measurement_fundamental_automaton2.average_speed * 1.0,
                        measurement_fundamental_automaton3.average_speed * 2.0,
                        measurement_fundamental_automaton4.average_speed * 3.0,
                        measurement_fundamental_automaton5.average_speed * 4.0])

densities_array = np.array([0.5, 1.0, 2.0, 3.0, 4.0])
```

```python
#plot fundamental diagram
plt.scatter(densities_array,flows_array)
plt.ylabel('Flow=Density*Speed')
plt.xlabel('Density')
plt.title('Flow depending on pedestrian density')
plt.savefig("results/task5_measurement_fundamental_diagramm.pdf")
```

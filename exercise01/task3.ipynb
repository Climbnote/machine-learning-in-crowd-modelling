```python
from automaton import Scenario
from automaton import Automaton
from automaton import Parameters
import visualization
import numpy as np
from enum import Enum
from IPython.display import HTML
import matplotlib.pyplot as plt
from matplotlib import colors
from matplotlib import animation
import math
import random
import scenario_reader
```

## Task 3/5: Interaction of pedestrians

### 1: Without Speed Control

```python
#scenario as described in the exercise sheet
circle_scenario = scenario_reader.make_scenario('scenarios/circle_scenario.json')

#plot the scenario
circle_plot = visualization.plot_automaton_states(circle_scenario.states)

#save plot
circle_plot.savefig("results/task3_circle_scenario.pdf")

#simulate the circle scenario using the Automaton class
circle_automaton = Automaton(circle_scenario)
circle_data = circle_automaton.simulate_pedestrians()

#animate the simulated data using the Visualization class

#set matplotlib environment
%matplotlib notebook
#call animation function and convert animation to JavaScript to be able to display it inline
circle_animation = visualization.plot_automaton_simulation(circle_data, Parameters(speed_pedestrian=1.0))
HTML(circle_animation.to_jshtml())
```

```python
#save the animation as a video file
circle_animation.save('results/task3_circle_scenario.mp4', codec='h264')
```

```python
#save scenario after 19 steps for report
circle_plot_timesteps_19 = visualization.plot_automaton_states(circle_data[19])

#save plot
circle_plot_timesteps_19.savefig("results/task3_circle_scenario_timestep_19.pdf")
```

### 2: With Speed Control

```python
#now we add a speed control to the automaton, such that pedestrians walk in roughly the same speed in arbitrary direction
circle_speed_scenario = scenario_reader.make_scenario('scenarios/circle_speed_scenario.json')

#simulate the circle scenario with speed control using the Automaton class
circle_speed_automaton = Automaton(circle_speed_scenario)
circle_speed_data = circle_speed_automaton.simulate_pedestrians()

#animate the simulated data using the Visualization class
#set matplotlib environment
%matplotlib notebook
#call animation function and convert animation to JavaScript to be able to display it inline
circle_speed_animation = visualization.plot_automaton_simulation(circle_speed_data, Parameters(speed_pedestrian=1.0))
HTML(circle_speed_animation.to_jshtml())
```

```python
#save the animation as a video file
circle_speed_animation.save('results/task3_circle_scenario_speed_control.mp4', codec='h264')
```

```python
#save scenario after 19 steps for report
circle_speed_plot_timesteps_19 = visualization.plot_automaton_states(circle_speed_data[19])

#save plot
circle_speed_plot_timesteps_19.savefig("results/task3_circle_scenario_speed_timestep_19.pdf")
```
